```python
data['Date/Time']= data['Date/Time'].map(pandas.to_datetime)   #convert Date/time column to pandas to date time 
                                                               #so that it can read 
```

```python
data['Date/Time'][0]                #reading first record of Date/time column
```

```python
#Function to return date of the months
```

```python
def get_dom(dt):                                               #function 1 : Day of the Month
    return dt.day
data ['Dom'] = data['Date/Time'].map(get_dom) 

def get_weekday(dt):                                           #function 2 : calculate weekday
    return dt.weekday()
data ['Weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):                                              #function 3 : Calculate hour
    return dt.hour
data ['Hour'] = data['Date/Time'].map(get_hour) 
```

# Analyses

### 1) Analyse DOM....

                     A)   ##Histogram Frequency By Date Of Month

```python
hist(data.Dom, bins=30, rwidth=0.8, range=(0.5,30.5))
xlabel('Date Of Months')
ylabel('Number of uber Trips')
title('Frequency By Date Of Month --- APRIL 2015')
```

                    B)    ##Histogram Frequency By Date Of Month (Sorted according to trips)

```python
##### BY For loop #########
#for k, rows in data.groupby('Dom'):            #group by all records according to days of month
#       print(k,len(rows))                      #that is day one is having 14546 trips 


##### By Function (more efficient) ####
def count_rows(rows):                           #function to count number of rows and return its length
    return len(rows)

by_date = data.groupby('Dom').apply(count_rows) #group by all records according to days of months and put them in a column 
```

```python
by_date_sorted= by_date.sort_values()                 #create a column which sort all the records by each date date


bar(range(1,31), by_date_sorted)
xticks(range(1,31),by_date_sorted.index)
xlabel('Date Of Months')
ylabel('Number of uber Trips')
title('Frequency By Date Of Month --- APRIL 2015')
```

### 2) Analyse hour

```python
hist(data.Hour,bins=24,range=(-0.5,24.5), rwidth=0.7,color='royalblue')
#data['Hour'].hist(bins = 24)
xlabel('Time of Day')
ylabel('Number of uber Trips')
title('Frequency By Time of the day --- APRIL 2015')
```

### 3) Analyse weekday

```python
hist(data.Weekday,bins=7,range=(-0.5,6.5),rwidth=0.7,color=("lightpink"),alpha=0.8)
xticks(range(7), 'Mon Tue Wed Thur Fri Sat Sun'.split())
xlabel('Week of the Day')
ylabel('Number of uber Trips')
title('Frequency By Week of the day --- APRIL 2015')
;
```
